```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time


driver = webdriver.Chrome()
base_url = "https://books.toscrape.com/catalogue/page-"

book_data = []


for page in range(1, 51):
    driver.get(f"{base_url}{page}.html")
    time.sleep(0.2)

    books = driver.find_elements(By.CSS_SELECTOR, "article.product_pod h3 a")
    book_links = [b.get_attribute("href") for b in books]

    for link in book_links:
        driver.get(link)
        time.sleep(0.1)

        title = driver.find_element(By.TAG_NAME, "h1").text
        price = driver.find_element(By.CSS_SELECTOR, "p.price_color").text
        availability = driver.find_element(By.CSS_SELECTOR, "p.availability").text.strip()

        book_data.append({
            "title": title,
            "price": price,
            "availability": availability
        })

driver.quit()


df = pd.DataFrame(book_data)


df["price"] = df["price"].str.replace("Â£", "").astype(float)
df["available_count"] = df["availability"].str.extract("(\d+)").astype(float)

print(df["available_count"].describe())
```
